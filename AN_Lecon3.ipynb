## 1. Méthode des rectangles

Cette méthode permet de calculer l’intégrale numérique en réalisant une somme de surfaces de rectangles. Le domaine d’intégration est découpé en intervalles et nous faisons comme si la fonction restait constante sur chaque intervalle.

Sur chaque intervalle, nous réalisons ainsi l’approximation suivante : $\int_a^b f(x) dx=(b-a) f(c)$ où $c$ est une abscisse appartenant à l’intervalle limité par $a$ et $b$.
Nous rappelons comment tracer des rectangles afin d'illustrer cette méthode. 

Afin de calculer une intégrale sur un intervalle $\lbrack a,b\rbrack$, nous considérons des sous-intervalles de même amplitude sur les quels nous choisissons une valeur prise par la fonction à l’extrémité gauche ou droite de l’intervalle (ou encore la valeur du milieu de l'intervalle).


```python
"""
Integration numérique par la 
méthode des rectangles avec c = a
pour la fonction f(x)=cos(x) sur 
[0,3pi/2]
"""
xmin = 0
xmax = 3*np.pi/2
nbx = 15
nbi = nbx - 1 # nombre d'intervalles

x = np.linspace(xmin, xmax, nbx)
print(x)
y = np.cos(x)
plt.plot(x,y,"bo-")

integrale = 0
for i in range(nbi):
    integrale = integrale + (y[i])*(x[i+1]-x[i])
    # dessin du rectangle
    x_rect = [x[i], x[i], x[i+1], x[i+1], x[i]] # abscisses des sommets
    y_rect = [0   , y[i], y[i]  , 0     , 0   ] # ordonnees des sommets
    plt.plot(x_rect, y_rect,"r")
print("integrale =", integrale)

plt.show()
```

# 2. Méthode des trapèzes

Cette méthode d’intégration utilise une somme de surfaces de trapèzes. Sur chaque intervalle, nous réalisons l’approximation suivante : $\int_a^b f(x) dx=\frac{b-a}{2}(f(a)+f(b))$. 
Pour calculer la surface du trapèze, nous sommons les aires du rectangle et du triangle rectangle qui le constituent. Nous obtenons que l'aire est égale à $\frac{b-a}{2}(f(a)+f(b))$.

```python
def dichotomie(f, a, b, epsilon):
    show(plot(f, a, b))
    if f(a)*f(b)>0:
        return("pas de solution sur l intervalle [", a, b, "]")
    else:
        N = floor((log(b-a)-log(epsilon))/log(2))+1
        print(N)
        n = 1
        while abs(b-a)>epsilon and n<N:
            c = (a+b)/2
            if f(a)*f(c)<0:
                b = c
            elif f(a)*f(c)>0:
                a = c        
            elif f(c)==0: 
                return("la solution est", c)
            n += 1
        return("la solution est", c.n(), "a", epsilon.n(), "pres", n)
```

```python
def pointfixe(g, x0, N):
    """
    Ici on va définir la suite x(n+1)=g(x(n))
    avec x(0) = x0
    """
    a = x0
    L = [a]
    for i in range(1,N+1):
        a = g(a).n()
        L += [a]
    # On retourne x(N)
    return(a, L)
```

On souhaite représenter graphiquement l'évolution d'une population de proies et de prédateurs évoluant suivant un système d'équation du type Lokta-Volterra :

$\frac{du}{dt} = au-buv$,
$\frac{dv}{dt}= -cv+duv$.

où $u$ désigne le nombre de proies (par ex. des lapins) et $v$ le nombre de prédateurs (par ex. des renards).
De plus, $a, b, c, d$ sont des paramètres qui décrivent l’évolution des populations : $a$ caractérise la croissance naturelle du nombre de lapins en l’absence de renards, $b$ la décroissance du nombre de lapins due à la présence de prédateurs, $c$ la décroissance du nombre de renards en l’absence de proies, enfin $d$ indique combien il faut de lapins pour qu’apparaisse un nouveau renard.

#### Markov Chain and Monte Carlo simulation

we want to know the probabilities for the current state for the next 20 steps when you started in S3. As an example, I'll use reproduction. The states are S1={AA,AA}, S2={AA,Aa}, S3={AA,aa}, S4={Aa,Aa}, S5={Aa,aa} and S6={aa,aa}.

The idea is that each pair of parents give birth to two children. The parents S2={AA,Aa} can give birth to {{AA, AA}, {AA, Aa}, {Aa, Aa}}. This results in the following state transition matrix.

The rows mean from which state you start, the colums are the states you can get to.



```python
#https://martin-thoma.com/python-markov-chain-packages/
import numpy as np
from matplotlib import pyplot

P = np.matrix([[1., 0., 0., 0., 0., 0.],
               [1./4, 1./2, 0., 1./4, 0., 0.],
               [0., 0., 0., 1., 0., 0.],
               [1./16, 1./4, 1./8, 1./4, 1./4, 1./16],
               [0., 0., 0., 1./4, 1./2, 1./4],
               [0., 0., 0., 0., 0., 1.]])

v = np.matrix([[0, 0, 1, 0, 0, 0]])

# Get the data
plot_data = []
for step in range(20):
    result = v * P**step
    plot_data.append(np.array(result).flatten())

# Convert the data format
plot_data = np.array(plot_data)

# Create the plot
pyplot.figure(1)
pyplot.xlabel('Steps')
pyplot.ylabel('Probability')
lines = []
for i, shape in zip(range(6), ['x', 'h', 'H', 's', '8', 'r+']):
    line, = pyplot.plot(plot_data[:, i], shape, label="S%i" % (i+1))
    lines.append(line)
pyplot.legend(handles=lines, loc=1)
pyplot.show()
```
